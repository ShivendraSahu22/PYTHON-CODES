```python
#PYTHON PROGRAM TO CHECK WHERE PYTHON PROGRAM TO CHECK WHERE A STRING IS PALINDROME OR NOT?#PYTHON PROGRAM TO CHECK WHERE PYTHON PROGRAM TO CHECK WHERE A STRING IS PALINDROME OR NOT?
s = input ("ENTER STRING TO CHECK")
s = s.replace(" ", "").lower()
if s == s[::-1]:
    print(s,"is a palindrome")
else:
    print(s,"is not a palindrome")
```
