Sum of natural number

```python
n = int(input("Enter n: "))
total = 0

for i in range(1, n+1):
    total += i

print("Sum =", total)
```

Leap year checker

```python
year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap year")
else:
    print("Not a leap year")
```
