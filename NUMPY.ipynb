FUNCTION TO CREATE ARRAY

```python
#using arange function
#arange is an array-valued version of the build-in python create arange function
a = np.arange(10)
a
```

```python
b = np.arange(1,2,3) #start,end (exclusive),step
b
```

```python
#using linspace ,leineare space
a = np.linspace(0,1,6)#start,end, number of points
a
```

```python
#common array
#1
a = np.ones((3,3))
a
```

```python
#common array
#0
a = np.zeros((3,3))
a
```

```python
c = np.eye(3) # Return a 2-D array with ones in the diagonal
c
```

```python
d = np.eye(3,2) #3 is number of rows, 2 is number of columns
d
```

```python

#create array using diag function
a = np.diag([1,2,3,4]) #construct a diagonal array
a
```

```python
np.diag(a) #Extract diagonal
```

```python
#Create array using random
#Create an array of the given shape and populate it with random
a = np.random.rand(4)
a
```

```python
a = np.random.randn(4)
a
```

```python
a = np.arange(10)
a.dtype
```

```python
#You can explicitly specify which data-type you want:
a = np.arange(10,dtype='float64')
a
```

```python
#The default data type is float zeros and ones function
a = np.zeros((3,3))
print(a)
a.dtype
```
