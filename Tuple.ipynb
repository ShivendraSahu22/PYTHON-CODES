Tuples

```

#empty
t = ()
#having
t = (1,2,3)
print(t)
#tuple with mixed datatype
t= (1,"name",2,"hii")
print(t)
#nested tuple
t = (1,(1,2,3),[1,"name",2,"hii"])
print(t)
```

Type of tuple

```
t = ('name')
print(type(t))
print(t)
t = ('name',)
print(type(t))
print(t)
#parenthesis is optional
t = 'name',
print(type(t))
print(t)
```

```
t = ("name","hii","hello","tata","bye")
print(t[2])
#negative index
print(t[-1])
```

```
#nested tuple
t = ('ABC',('hii','hello','bye'))
print(t[1])
print(t[1][2])
```
