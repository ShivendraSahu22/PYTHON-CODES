```python
a = np.array([[5,6,4],[4,3,4]])
b = np.sort(a,axis=1)
b
```

```python

a = np.array([[5,6,4],[4,3,4]])
b = np.sort(a,axis=0)
b
```

```python
a = np.array([[5,6,4],[4,3,4]])
b = np.sort(a,axis=1)[:, ::-1] # Sorts in descending order
b
```

```python
a = np.array([[5,6,4],[4,3,4]])
b = np.sort(a,axis=1)[:, ::1] # Sorts in ascending order
b
```
