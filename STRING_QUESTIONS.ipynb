count the number of vowels in string,

```python
count = 0
s = input("ENTER THE NAME")
for l in s:
    if l in 'aeiouAEIOU':
        count += 1
print(count,'letter found')
```

Count Words

```python
s = input("ENTER THE STRING: ")
words = s.split()
count = len(words)
print('words found', count,'times')
```

Reverse String

```python
s = input("Enter a string: ")
reversed_s = s[::-1]
print("Reversed string:", reversed_s)
```

Find Character Frequency

```python
s = input("ENTER NAME")
a = input("ENTER TO FIND CHAR FREQUENCY")
count = 0
for l in s:
    if l == a:
        count += 1
print(a,'char found',count,'times')
```

Check anagram

```python
def is_anagram(str1, str2):
    return sorted(str1.lower()) == sorted(str2.lower())

# Get user input for the strings
st1 = input("Enter the first string: ")
st2 = input("Enter the second string: ")

# Test with user input
if is_anagram(st1,st2):
    print( {st1} and {st2}, "are anagram.")
else:
    print( {st1} and {st2}, "are not anagram.")
```
